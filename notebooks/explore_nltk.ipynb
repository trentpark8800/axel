```python
from random import shuffle
import nltk
from nltk.corpus import twitter_samples
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd

# Set display options
pd.set_option("display.max_colwidth", None)
pd.set_option("display.max_columns", None)
```

```python
schema = {
    "created_at": "object",
    "tweet_id": "int64",
    "tweet": "object",
    "retweets": "int16"
}
```

```python
tweets_df = pd.read_csv("./data/raw/standard_bank_tweets.csv", header=0, dtype=schema)
tweets_df.head(5)
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
# Clean links from the tweet column
tweets_df['tweet'] = tweets_df['tweet'].str.replace("://", "//")
tweets_df.head(5)
```

```python
tweets_df['sentiment_scores'] = tweets_df['tweet'].apply(lambda tweet: sia.polarity_scores(tweet))
tweets_df.head(10)
```

```python
tweets_df['compound_sentiment'] = tweets_df['sentiment_scores'].apply(lambda scores: scores['compound'])
tweets_df.tail(20)
```

```python
# Let's look at negative tweets
tweets_df[tweets_df['compound_sentiment'] < 0].head(5)
```
